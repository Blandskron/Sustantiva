#Graficos

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Creamos un conjunto de valores x desde -10 hasta 10

```python
x = np.linspace(-10, 10, 100)
```

# Calculamos los valores y correspondientes a y = x^2

```python
y = x ** 2
```

# Creamos el gr√°fico

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x^2', color='blue', linewidth=2)
```

# A√±adimos etiquetas y t√≠tulo

```python
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.title('Gr√°fico de y = x^2', fontsize=16)
```

# A√±adimos una leyenda

```python
plt.legend()
```

# Mostramos el gr√°fico

```python
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Creamos un conjunto de valores x desde -10 hasta 10
x = np.linspace(-10, 10, 100)

# Calculamos los valores y correspondientes a y = x^2
y = x ** 2

# Creamos el gr√°fico
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x^2', color='blue', linewidth=2)

# A√±adimos etiquetas y t√≠tulo
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.title('Gr√°fico de y = x^2', fontsize=16)

# A√±adimos una leyenda
plt.legend()

# Mostramos el gr√°fico
plt.grid(True)
plt.show()
```
