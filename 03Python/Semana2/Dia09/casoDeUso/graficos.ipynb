#Graficos

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Creamos un conjunto de valores x desde -10 hasta 10

```python
x = np.linspace(-10, 10, 100)
```

# Calculamos los valores y correspondientes a y = x^2

```python
y = x ** 2
```

# Creamos el gráfico

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x^2', color='blue', linewidth=2)
```

# Añadimos etiquetas y título

```python
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.title('Gráfico de y = x^2', fontsize=16)
```

# Añadimos una leyenda

```python
plt.legend()
```

# Mostramos el gráfico

```python
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Creamos un conjunto de valores x desde -10 hasta 10
x = np.linspace(-10, 10, 100)

# Calculamos los valores y correspondientes a y = x^2
y = x ** 2

# Creamos el gráfico
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x^2', color='blue', linewidth=2)

# Añadimos etiquetas y título
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.title('Gráfico de y = x^2', fontsize=16)

# Añadimos una leyenda
plt.legend()

# Mostramos el gráfico
plt.grid(True)
plt.show()
```
